### Estadísticos de variable de interés





### Nacionalidad



```python
nacionalidad = pd.Categorical(datos['Nationality'])
nacionalidad.describe()
```

```python
frecuencia_nac = (datos.groupby("Nationality").agg(frequency=("Nationality", "count")))
type(frecuencia_nac)
# frecuencia_nac
frecuencia_nac = frecuencia_nac.sort_values(['frequency'], ascending=False)
frecuencia_nac_topten = frecuencia_nac.head(10)
frecuencia_nac_topten

```

```python
frecuencia_nac_topten.plot(kind="bar")
```

### Pié preferido

```python
pie_preferido = pd.Categorical(datos['Preferred.Foot'])
pie_preferido.describe()
```

```python
frecuencia_pie = (datos.groupby("Preferred.Foot").agg(frequency=("Preferred.Foot", "count")))
type(frecuencia_pie)
# frecuencia_nac
frecuencia_pie = frecuencia_pie.sort_values(['frequency'], ascending=False)
frecuencia_pie
```

```python
frecuencia_pie.plot(kind="bar")
```
